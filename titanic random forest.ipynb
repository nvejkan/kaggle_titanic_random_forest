Let's make it a little bit more complicated...
Show no. of passengers and survival group by sex and pclass.

```python
l_sex_class = []

for i in range(df['Pclass'].min(),df['Pclass'].max()+1):
    for sex in ['male','female']:
        n_passenger = (df[ (df['Pclass'] == i) & (df['Sex'] == sex) ]['Survived']).count()
        survival = (df[ (df['Pclass'] == i) & (df['Sex'] == sex) ]['Survived']).sum()
        print('Sex: {3},\t Class: {0}, people: {1}, survivor: {2}'.format(i,n_passenger,survival,sex))
        l_sex_class.append((i,sex,n_passenger,survival))
```

```python
l_sex_class
```

Plot data from the l_sex_class list

```python
import matplotlib.pyplot as plt

n_groups = len(l_sex_class)

passenger = [ i[2] for i in l_sex_class ]
survivor = [ i[3] for i in l_sex_class ]
ticks = [ '({0},{1})'.format(i[0],i[1]) for i in l_sex_class ]

fig, ax = plt.subplots()
bar_width = 0.35

index = [ i for i in range(0,n_groups)]
index2 = [ i+bar_width for i in range(0,n_groups)]



rects1 = plt.bar(index, passenger, bar_width,
                 color='b',
                 label='passenger')

rects2 = plt.bar(index2, survivor, bar_width,
                 color='g',
                 label='survivor')

plt.xlabel('Group')
plt.ylabel('no. of passengers')
plt.title('Survivor by sex and class')
plt.xticks(index2, ticks)
plt.legend() # show legend on the top-left

plt.tight_layout()
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```

# Start cleaning data

```python
df['Gender'] = 1 # Adding new columns
df.head(3)
```

```python
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df.head(3)
```

```python
import numpy as np
median_ages = np.zeros((2,3)) # create a 2*3 array; 2 sexes, 3 classes

for i in range(0,2): #sex
    for j in range(0,3): #pclass
        median_ages[i,j] = df[ (df['Gender'] == i) & (df['Pclass'] == j+1) ]['Age'].dropna().median()
        
median_ages
```

```python
df['AgeFill'] = df['Age'] #new col
for i in range(0,2): #sex
    for j in range(0,3): #pclass
        df.loc[ (df['Gender'] == i) & (df['Pclass'] == j+1) & (df['Age'].isnull()) ,'AgeFill'] = median_ages[i,j]  #loc[row,col]

df[ df['Age'].isnull() ].head(5)
```

```python
df['AgeIsNull'] = pd.isnull(df.Age).astype(int) # another new col 0 = not null, 1 = null
df.head(6)
```

```python
df.describe()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['Age*Class'] = df.AgeFill * df.Pclass

#The next step is to drop the columns which we will not use (not numbers)
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked','Age'], axis=1) # axis 1 is col

#drop any rows which still have missing values
df = df.dropna()
```
